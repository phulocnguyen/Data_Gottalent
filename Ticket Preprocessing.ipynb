```python
import numpy as np
import pandas as pd
```

```python
excel_file_path = 'Data/dataset.xlsx'

sheet_name = 'Ticket'  

ticket_data = pd.read_excel(excel_file_path, sheet_name=sheet_name)

df_ticket = pd.DataFrame(ticket_data)
```

```python
df_ticket.dropna(inplace = True)
```

```python
df_ticket['saledate'] = pd.to_datetime(df_ticket['saledate'])
df_ticket['date'] = pd.to_datetime(df_ticket['date'])
```

```python
df_ticket['time_diff'] = (df_ticket['saledate'] - df_ticket['date'])/pd.Timedelta(hours=1)
```
